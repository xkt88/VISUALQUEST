```python
import os

def get_image_files(folder_path):
    # List to store image filenames
    image_files = []
    
    # Common image file extensions
    image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp')
    
    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file has an image extension
        if filename.lower().endswith(image_extensions):
            image_files.append(filename)
    
    return image_files

# Example usage
folder_path = 'celeb'  # Replace with your folder path
image_list = get_image_files(folder_path)
# print(image_list)
print(len(image_list))
#print(image_list)
for img in image_list:
    if '.jpg' not in img:
        print(img)
```

```python
import os
from imghdr import what

# Replace this with the path to your folder
folder_path = "img"

def check_non_jpg_images(folder_path):
    non_jpg_files = []

    # Iterate through all files in the folder
    for root, _, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)

            # Check the file extension
            if not file.lower().endswith(".jpg"):
                # Further validate if it's an image
                detected_format = what(file_path)

                # If it's an image but not jpg, add to the list
                if detected_format and detected_format != "jpeg":
                    non_jpg_files.append((file, detected_format))

    return non_jpg_files

# Run the function
non_jpg_files = check_non_jpg_images(folder_path)

if non_jpg_files:
    print("The following files are not JPG format:")
    for file, format in non_jpg_files:
        print(f"- {file}: Detected format is {format}")
else:
    print("All files are in JPG format.")
```
